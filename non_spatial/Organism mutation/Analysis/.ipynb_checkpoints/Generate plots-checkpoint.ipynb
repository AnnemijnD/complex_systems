```python
import numpy
import pickle 
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter
```

```python
with open('cdict5000gen.pkl', 'rb') as f:
    cdict = pickle.load(f)
```

```python
# cdict = {k: v for k, v in sorted(cdict.items(), key=lambda item: item[1])}
cdict.values()
```

```python

c = 0
for name, age in cdict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
    if age == 818781:
        pat = age
        c += 1
print(c)
```

```python
for ci in range(900,1000):
    print(ci)
    c=0
    for name, age in cdict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
        if age == 1:
            pat = name
            if c == ci:
                break
            c += 1
    pattern = [pat[i*9:i*9+9] for i in range(5)]
    sns.heatmap(pattern)
    plt.show()
```

```python
c = 0
for name, age in cdict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
    if age == 1:
        pat = name
        if c == 948:
            break
        c += 1
pattern = [pat[i*9:i*9+9] for i in range(5)]
sns.heatmap(pattern)
with open('complexpattern.pkl', 'wb') as f:
    pickle.dump(pattern, f)
plt.show()
```
